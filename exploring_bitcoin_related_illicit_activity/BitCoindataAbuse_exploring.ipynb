Checking the common address of CryoptoScamDB and Kaggle Scam Dataset

```python
crypto_df = df3[['address']].copy()
kaggle_df = df_[['Address']].copy()
crypto_df.rename(columns = {'address':'Address'}, inplace = True)
crypto_df=crypto_df.drop_duplicates(keep='first')
kaggle_df=kaggle_df.drop_duplicates(keep='first')
print(crypto_df.shape)
print(kaggle_df.shape)
```

```python
crypto_df=crypto_df.reset_index(drop=True)
kaggle_df=kaggle_df.reset_index(drop=True)
```

```python
display(crypto_df)
```

```python
display(kaggle_df)
```

```python
crypto_df.info()
kaggle_df.info()
print(crypto_df.shape)
print(kaggle_df.shape)
print(len(crypto_df.Address.unique()))
print(len(kaggle_df.Address.unique()))
```

```python
resData = crypto_df.merge(kaggle_df, on="Address", how = 'inner' ,indicator=False)
resData.tail()
```

```python
print("Common Addresses of CryptoScamDB and Kaggle Dataset",resData.shape)
```

```python
df_all_rows = pd.concat([crypto_df, kaggle_df],join='inner',ignore_index=True)
print(df_all_rows.shape)
df_all_rows.drop_duplicates(keep='first',inplace=True)
print(df_all_rows)
```

```python
display(df_all_rows)
```

```python
df_all_rows.to_csv('MargedAddresses.csv')
```
